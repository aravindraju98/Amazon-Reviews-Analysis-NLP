```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from textblob import TextBlob
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_json('Software.json.gz',lines=True,compression='gzip')
```

```python
df = df.dropna(subset=['reviewText'])
```

```python
df=df.drop(["verified","reviewTime","reviewerID","style","reviewerName","overall","unixReviewTime","vote","image", "summary" ], axis = 1)
```

```python
df.asin.value_counts()
```

```python
df_filtered = df[df['asin'] == 'B00UB76290']
```

```python
df_filtered.head()
```

```python
df_filtered.shape
```

```python
df_filter_lvl2= df_filtered.drop(["asin"], axis=1)
```

```python
df_filter_lvl2
```

```python

reviews = df_filter_lvl2['reviewText']

# Perform Sentiment Analysis
def analyze_sentiment(review):
    analysis = TextBlob(review)
    return analysis.sentiment.polarity 
```

```python
df_filter_lvl2['sentiment_score'] = reviews.apply(analyze_sentiment)

# Categorize Sentiment
def categorize_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
df_filter_lvl2['sentiment_category'] = df_filter_lvl2['sentiment_score'].apply(categorize_sentiment)

# Visualize Sentiment Distribution
plt.figure(figsize=(8, 5))
df_filter_lvl2['sentiment_category'].value_counts().plot(kind='bar', color=['green', 'red', 'gray'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
df_filter_lvl2.head()
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    tokens = word_tokenize(text.lower())
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word.isalpha() and word not in stop_words]
    return " ".join(tokens)

reviews_clean = df_filter_lvl2['reviewText'].apply(preprocess)
```

```python
reviews_clean.head()
```

```python
vectorizer = CountVectorizer(max_df=0.95, min_df=1, stop_words='english')
dtm = vectorizer.fit_transform(reviews_clean)
```

```python
# Step 3: Train LDA Model
lda = LatentDirichletAllocation(n_components=5, random_state=42)  # Adjust n_components as needed
lda.fit(dtm)
```

```python
# Step 4: Display Topics
def display_topics(model, feature_names, num_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[-num_top_words:]]))

display_topics(lda, vectorizer.get_feature_names_out(), 10)
```

```python
# Step 5: Assign Topics to Reviews
topics = lda.transform(dtm)
dominant_topic = topics.argmax(axis=1)
```

```python
# Step 6: Analyze Sentiment-Topic Relationship
df_filter_lvl2['Dominant_Topic'] = dominant_topic
sentiment_topic_pivot = df_filter_lvl2.groupby(['Dominant_Topic', 'sentiment_category']).size().unstack()

# Step 7: Visualize Insights
sentiment_topic_pivot.plot(kind='bar', stacked=True)
plt.title("Sentiment Distribution Across Topics")
plt.show()
```
